##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 6, then the output should be `[36, 25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
# Write your code here

import numpy as np
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt

def generator(x):
  arr=[0]*x
  val=x
  while(val>0):
    arr[x-val]=val*val
    val=val-1
  return arr


print (generator(6))
print (generator(10))
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
# Write your code here
import numpy as np
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt

X= np.arange(start = -3, stop = 3, step = 0.5)
Y= np.arange(start = -3, stop = 3, step = 0.5)
X,Y=np.meshgrid(X,Y)

Z= 1 - X - Y

fig=plt.figure()
ax =plt.axes(projection="3d")

ax.plot_surface(X,Y,Z, cmap="viridis", edgecolor="red")
ax.set_title('3D Graph')

#Extras
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_zlabel("z-axis")

plt.show()

```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
# Write your code here

import numpy as np
import matplotlib.pyplot as plt

# Define the function to calculate the Taylor approximation of e^x
def taylor_approximation(x, order):
    return 1 + x + (x**2 / 2) if order >= 2 else 1 + x

# Define the function to calculate the mean square error (MSE)
def mse(original, approximation):
    return np.mean((original - approximation)**2)

# Define the range and interval for x values
x_values = np.arange(start=-1,stop= 1.01,step= 0.01)

# Calculate the original functions
e2x = np.exp(2 * x_values)
e3x = np.exp(3 * x_values)
e5x = np.exp(5 * x_values)

# Calculate the Taylor approximations for each method
taylor_approximation_1 = taylor_approximation(x_values, 2) * taylor_approximation(x_values, 2)
taylor_approximation_2 = taylor_approximation(x_values, 5)

mse_1 = mse(e2x, taylor_approximation_1)
mse_2 = mse(e5x, taylor_approximation_2)

print("MSE for Method 1:", mse_1)
print("MSE for Method 2:", mse_2)

plt.plot(x_values, e2x, label="e^(2x)")
plt.plot(x_values, e3x, label="e^(3x)")
plt.plot(x_values, e5x, label="e^(5x)")
plt.plot(x_values, taylor_approximation_1, label="Taylor Approximation (Method 1)")
plt.plot(x_values, taylor_approximation_2, label="Taylor Approximation (Method 2)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("The Graph")
plt.legend()
plt.show()
```
